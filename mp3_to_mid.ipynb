```python
VOICE_DIR = '/content/drive/MyDrive/DataSetVoice/voice'
GUITAR_DIR = '/content/drive/MyDrive/DataSetVoice/guitar_song'
VOICE_WAV_DIR = '/content/drive/MyDrive/DataSetVoice/voice_wav'
GUITAR_WAV_DIR = '/content/drive/MyDrive/DataSetVoice/guitar_wav'
VOICE_MIDI_DIR = '/content/drive/MyDrive/DataSetVoice/voice_midi'
GUITAR_MIDI_DIR = '/content/drive/MyDrive/DataSetVoice/guitar_midi'
```

```python
import os
import pydub
import subprocess
import audio_to_midi
```

```python
voice_list = os.listdir(VOICE_DIR)
guitar_list = os.listdir(GUITAR_DIR)

print(len(voice_list))
print(len(guitar_list))
```

```python
guitar_list_match = [name.split('_')[1] for name in guitar_list]
matched_voice_list = [voice for voice in voice_list if voice in guitar_list_match]
len(matched_voice_list)
```

```python
matched_voice_list.sort()
guitar_list.sort()
```

## Convert MP3 to WAV

```python
from pydub import AudioSegment

for song in matched_voice_list:
  sound = AudioSegment.from_mp3(os.path.join(VOICE_DIR, song))
  sound.export(os.path.join(VOICE_WAV_DIR, song.split('.')[0]+'.wav'), 
               format="wav")
```

```python
for song in guitar_list:
  sound = AudioSegment.from_mp3(os.path.join(GUITAR_DIR, song))
  sound.export(os.path.join(GUITAR_WAV_DIR, song.split('.')[0]+'.wav'), 
               format="wav")
```
