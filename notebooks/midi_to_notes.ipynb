```python
import os
import librosa
```

```python
HOME_DIR = '/home/kolawolesteven99/Downloads'
VOICE_MIDI_DIR = os.path.join(HOME_DIR, 'voice_midi')
GUITAR_MIDI_DIR = os.path.join(HOME_DIR, 'guitar_midi')
```

```python
voice_list = os.listdir(VOICE_MIDI_DIR)
guitar_list = os.listdir(GUITAR_MIDI_DIR)

print(len(voice_list))
print(len(guitar_list))
```

```python
voice_list.sort()
guitar_list.sort()
```

## Midi to Notes

```python
!pip install mido
```

```python
from tqdm import tqdm
from mido import MidiFile
```

### Converting a song

```python
os.chdir(VOICE_MIDI_DIR)
for song in voice_list[:1]:
  single_song = []
  midi = MidiFile(song)
  for i, track in enumerate(tqdm(midi.tracks)):
    print('Track {}: {}'.format(i, track.name))
    for line in track:
      note = librosa.midi_to_note(line.bytes())
      single_song.append(note)
      print(line.bytes(), note)
    print()
  print(len(single_song))
```

### Converting a Guitar

```python
os.chdir(GUITAR_MIDI_DIR)
for guitar in guitar_list[:1]:
  single_guitar = []
  midi = MidiFile(guitar)
  for i, track in enumerate(tqdm(midi.tracks)):
    print('Track {}: {}'.format(i, track.name))
    for line in track:
      note = librosa.midi_to_note(line.bytes())
      single_guitar.append(note)
      print(line.bytes(), note)
    print()
  print(len(single_guitar))
```

### Full-scale Conversion

```python
os.chdir(VOICE_MIDI_DIR)
song_notes = []

for song in tqdm(voice_list):
  single_song = []
  midi = MidiFile(song)
  for track in midi.tracks:
    for line in track:
      note = librosa.midi_to_note(line.bytes())
      single_song.append(note)
  song_notes.append(single_song)
  
print("Total number of songs with notes:", len(song_notes))
```

```python
os.chdir(GUITAR_MIDI_DIR)
guitar_notes = []

for guitar in tqdm(guitar_list):
  single_guitar = []
  midi = MidiFile(guitar)
  for track in midi.tracks:
    for line in track:
      note = librosa.midi_to_note(line.bytes())
      single_guitar.append(note)
  guitar_notes.append(single_guitar)
print("Total_number of guitar with notes", len(guitar_notes))
```

## Save in CSV files

```python
import csv

os.chdir(HOME_DIR)
with open("song_notes.csv", "w", newline="") as f:
  writer = csv.writer(f)
  writer.writerows(song_notes)
```

```python
with open("guitar_notes.csv", "w", newline="") as f:
  writer = csv.writer(f)
  writer.writerows(guitar_notes)
```
